```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["python","javaScript","java","c++"]
y = np.random.randint(200,size=4)
y2 = np.random.randint(100,size=4)
plt.figure(figsize=(10,3))
plt.title("bar Graph",fontsize=20)
plt.xlabel("languages",fontsize=20)
plt.ylabel("ranks",fontsize=20)
c= ["r","g","b","m"]
ec = c.copy()
ec.reverse()

plt.bar(x,y,width=0.2,align="edge",label="popularity1",color=c,linewidth=2,edgecolor=ec,alpha=0.7)

plt.bar(x,y2,width=0.2,align="edge",label="popularity2",color=c,linewidth=2,edgecolor=ec)
plt.legend()
plt.show()
```

```python
x = ["python","javaScript","java","c++"]
y = np.random.randint(200,size=4)
y2 = np.random.randint(100,size=4)
width = 0.3
x_indexes = np.arange(len(x))
plt.figure(figsize=(10,3))
plt.title("bar Graph",fontsize=20)
plt.xlabel("languages",fontsize=20)
plt.ylabel("ranks",fontsize=20)
c= ["r","g","b","m"]
ec = c.copy()
ec.reverse()

plt.bar(x_indexes-width,y,width=width,align="edge",label="popularity1",color=c,linewidth=2,edgecolor=ec,alpha=0.7)

plt.bar(x_indexes,y2,width=width,align="edge",label="popularity2",color=c,linewidth=2,edgecolor=ec)
plt.xticks(ticks=x_indexes,labels=x,rotation=20)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_ages = [20,33,44,55,60,70]
y_dev_salary = [102309,222309,402223,500504,300200,223098]
a =np.array(y_dev_salary)
p_dev_salary = a+100000
j_dev_salary = a+63021
j_dev_salary[2]=502345
j_dev_salary[5]=102345
# print(plt.style.available)
plt.figure(figsize=(10,7))
plt.style.use('classic')
plt.plot(x_ages,y_dev_salary,color='k' ,label="all dev",linewidth=5,marker='.',linestyle='--')
plt.plot(x_ages,p_dev_salary,color='m' ,label="python dev",linewidth=3,marker='o')
plt.plot(x_ages,j_dev_salary,color='m' ,label="javaScript dev",linewidth=3,marker='o')

plt.legend()
plt.title("developers salary")
plt.xlabel("ages")
plt.ylabel("salary")
# plt.grid()
plt.tight_layout()
plt.savefig("plot.png")
plt.show()
```
